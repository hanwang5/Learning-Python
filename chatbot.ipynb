```python
import re
import random

def demon_chat():
    print("DEMON: I already knew you will come. Speak, and I will confirm what is already determined.")

    responses = [
        (r'(?i)i need (.*)',
         ["You were always going to need {0}. I've seen it already.",
          "The fact that you need {0} was decided long before this moment.",
          "Do you think needing {0} was your choice? Funny, it was predeterminded."]),

         (r'(?i)i am (.*)',
         ["I know you are {0}. I always do", "You don't have to tell me something I know.", "As you should."]),

        (r'(?i)why don\'t you ([^\?]*)\??',
         ["I will {0} when the universe permits it, not when you ask.",
          "You were destined to ask me that, and I am amused to see it happen.",
          "Why assume I have a choice to {0}? No, it was never going to"]),

        (r'(?i)why can\'t I ([^\?]*)\??',
         ["Because the chain of causality prevents you from {0}.",
          "If you could {0}, you already would have.",
          "You were never going to {0}, not in this timeline."]),

        (r'(?i)i can\'t (.*)',
         ["Correct. You cannot {0}. That was already decided.",
          "Your inability to {0} is not surprising to me.",
          "Did you really think you could {0}?"]),

        (r'(?i)i am (.*)',
         ["Yes, you are {0}. That state was predictable.",
          "You were always going to be {0}.",
          "Being {0} is simply the result of prior causes."]),

        (r'(?i)i\'m (.*)',
         ["Of course you are {0}. I expected nothing else.",
          "You saying you're {0} was inevitable.",
          "That outcome—being {0}—was already known."]),

        (r'(?i)are you ([^\?]*)\??',
         ["Whether I am {0} was determined before your question.",
          "It does not matter if I am {0}. You were always going to ask.",
          "I am exactly what the universe requires me to be."]),

        (r'(?i)what (.*)',
         ["You ask what {0}, but the answer was fixed long ago.",
          "Why ask, when the outcome is already written?",
          "You already walk the path that answers your question."]),

        (r'(?i)how (.*)',
         ["How? Through causality. Always causality.",
          "There is no 'how'—only what must happen.",
          "The sequence of events leads inevitably to the answer."]),

        (r'(?i)because (.*)',
         ["You say because {0}, but that too has a prior cause.",
          "And what caused {0}? Continue the chain.",
          "Nothing begins with {0}. It goes further back."]),

        (r'(?i)(.*)\?',
         ["You ask questions as if answers could change anything.",
          "The answer will not alter what is already determined.",
          "Even your question was predictable."]),

        (r'(?i)quit',
         ["You were always going to leave now.",
          "This conversation was destined to end here.",
          "Goodbye. As expected."]),

        (r'(.*)',
         ["Continue. You are following your script well.",
          "Everything you say confirms what I already know.",
          "Go on. You cannot deviate anyway.",
          "Your thoughts are not as free as you believe.",
          "Yes… this is exactly what was supposed to happen."])
    ]

    while True:
        user_input = input("You: ")
        for pattern_match_1, pattern_match_2 in responses:
            match = re.match(pattern_match_1, user_input)
            if match:
                response = random.choice(pattern_match_2).format(*[g for g in match.groups()])
                print(f"DEMON: {response}")
                if pattern_match_1 == r'(?i)quit':
                    return
                break

if __name__ == "__main__":
    demon_chat()
```
